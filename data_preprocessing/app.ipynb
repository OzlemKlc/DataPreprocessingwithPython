```python
!pip install scipy
from scipy.stats import chi2_contingency
# Assuming your dataset is stored in a pandas DataFrame called 'df'
features = df.drop(['Machine failure','UDI','Product ID','HDF','OSF','PWF','TWF','RNF'], axis=1)  # Exclude the target variable
target = df['Machine failure']
chi2_scores, p_values = [], []
for feature in features.columns:
    contingency_table = pd.crosstab(features[feature], target)
    chi2, p, _, _ = chi2_contingency(contingency_table)
    chi2_scores.append(chi2)
    p_values.append(p)
result = pd.DataFrame({'Feature': features.columns, 'Chi-Square': chi2_scores, 'p-value': p_values})
result = result.sort_values(by='Chi-Square', ascending=False)  # or by 'p-value'
print(result)
```

```python
import pandas as pd
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import LabelEncoder


# Kategorik sütunları seçin
categorical_columns = ['Type']

# Label Encoding işlemi
label_encoder = LabelEncoder()
for column in categorical_columns:
    df[column] = label_encoder.fit_transform(df[column])

# Dummy Encoding işlemi
#df = pd.get_dummies(df, columns=categorical_columns)

df= df.select_dtypes(include=['float64','int64' , 'int32'] )
# Özellikleri ve hedef değişkeni ayırın
X = df.drop('Machine failure', axis=1)
y = df['Machine failure']

# SMOTE ile veri kümesini dengesizliği düzeltin
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Düzeltmiş veri kümesini yeni bir DataFrame'e dönüştürün
df_resampled = pd.DataFrame(X_resampled, columns=X.columns)
df_resampled['Machine failure'] = y_resampled


# Sonuçları gösterin
print("Düzeltmiş Veri Kümesi:")
df=df_resampled.copy()
print(df_resampled.tail())

```

```python
!pip install flask
# !pip install --upgrade Flask
from flask import Flask, render_template, request
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from flask import Flask, render_template, request
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from keras.models import load_model


# LSTM modelini oluşturun
model = Sequential([
    LSTM(64, activation='relu', input_shape=(X_train.shape[1], 1)),
    Dense(1, activation='sigmoid')
])

# Modeli derleyin
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# LSTM modeline uygun şekilde veri kümesini yeniden şekillendirin
# X_train= X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
# X_test= X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

# Modeli eğitin
model.fit(X_train, y_train, epochs=10, batch_size=32)
# int_features=[0, 300.499654	, 309.900000, 1396.994807, 45.901385,210]
# temp = np.array([int_features])
# b = model.predict(temp)
# print(b)

app = Flask(__name__)


@app.route('/' , methods=['GET'])
def index():
    return render_template('index.html')

@app.route('/predict', methods=['POST'])
def predict():
    # Kullanıcıdan giriş verilerini alın
    Product_type = float(request.form['type'])
    air_temp = float(request.form['air_temp'])
    process_temp = float(request.form['process_temp'])
    rot_speed = float(request.form['rot_speed'])
    torque = float(request.form['torque'])
   
    tool_wear = float(request.form['tool_wear'])
    
    
    input_data = np.array([[Product_type,air_temp, process_temp, rot_speed, torque, tool_wear]])
    input_df = pd.DataFrame(input_data, columns=['Type','Air temperature [K]', 'Process temperature [K]', 'Rotational speed [rpm]', 'Torque [Nm]', 'Tool wear [min]'])
    
   
    # Ölçekleme işlemini gerçekleştirin
    input_scaled = scaler.transform(input_df)

    # Tahmin yap
    prediction = model.predict(input_scaled)
    print(prediction)
    # Tahmin sonucunu göstermek için bir mesaj oluştur
    if prediction[0, 0] >= 0.5:
        result_message = f"Machine failure risk is HIGH."
    else:
        result_message = f"Machine failure risk is LOW."
    
    return render_template('result.html', result_message=result_message)

if __name__ == '__main__':
    app.run()


# # Modeli test edin
# loss, accuracy = model.evaluate(X_test, y_test)
# print('Test loss:', loss)
# print('Test accuracy:', accuracy)
```
